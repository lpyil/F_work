### import

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize
plt.figure(figsize=(5, 3), dpi=100)

# plt.plot(x, y, label = '2x', color = 'red', linewidth = 3, linestyle = '--',
# marker = '.', markersize = 10, markeredgecolor = 'green')

# can be use shortcuts
plt.plot(x, y, 'bo--', label='2x')  # fmt = '[color][marker][line]'
# line 2

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('First', fontdict={'fontname': 'Arial', 'fontsize': 15})
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks([0, 1, 2, 3, 4])
#plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()  # parametres

plt.savefig('mygraph.png', dpi=300)  # save
plt.show()
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)


bars[0].set_hatch('.')
bars[1].set_hatch('/')
bars[2].set_hatch('o')
# or

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6, 4))
plt.show()
```
